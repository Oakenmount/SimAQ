```python
from simaq import Simaq

model = Simaq(config="configs/default.yaml",
              weights="checkpoints/vital_yogurt_270_weights.pth",
              device="cuda:0")
```

```python
from tifffile import imread, imwrite
import glob

for fpath in glob.glob("../test_data/*.tif"):
    # if ends with _pred.tif, skip
    if fpath.endswith("_pred.tif"):
        continue
    
    # if version with prediction exists, skip
    if fpath.replace(".tif", "_pred.tif") in glob.glob("../test_data/*_pred.tif"):
        continue

    print(f"Processing {fpath}...")
    img = imread(fpath)
    pred = model.predict(img, normalize=True)
    imwrite(fpath.replace(".tif", "_pred.tif"), pred)
```
